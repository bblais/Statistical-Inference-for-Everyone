## Estimating Lengths

### Known deviation, $\sigma$

```python
x=[5.1, 4.9, 4.7, 4.9, 5.0]
sigma=0.5
```

```python
mu=sample_mean(x)
N=len(x)
```

```python
dist=normal(mu,sigma/sqrt(N))
distplot(dist)
```

```python
credible_interval(dist)
```

### Unknown $\sigma$

```python
mu=sample_mean(x)
s=sample_deviation(x)
print mu,s
```

```python
dist=tdist(N-1,mu,s/sqrt(N))
```

```python
distplot(dist,xlim=[4.6,5.4])
```

```python
credible_interval(dist)
```
